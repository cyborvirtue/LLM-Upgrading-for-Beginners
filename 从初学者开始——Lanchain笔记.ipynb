```python
import os
from typing import Any, List, Optional, Dict
from langchain.llms.base import LLM
from langchain.callbacks.manager import CallbackManagerForLLMRun
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from openai import OpenAI
# è®¾ç½®APIå¯†é’¥ï¼Œosæ˜¯è®¾ç½®ç¯å¢ƒå˜é‡
os.environ["Doubao_API_KEY"] = 'your_api_key'

# è‡ªå®šä¹‰è±†åŒ… LLM ç±»
class DoubaoLLM(LLM):
    model_name: str = "doubao-1-5-lite-32k-250115"
    
    def __init__(self):
        super().__init__()
        self._client = OpenAI(
            base_url="https://ark.cn-beijing.volces.com/api/v3",
            api_key=os.environ.get("Doubao_API_KEY")
        )
    
    @property
    def client(self) -> Any:
        return self._client

    @client.setter
    def client(self, value: Any):
        self._client = value

    def _call(self, prompt: str, **kwargs) -> str:
        completion = self.client.chat.completions.create(
            model=self.model_name,
            messages=[
                {"role": "user", "content": prompt}
            ]
        )
        return completion.choices[0].message.content

    @property
    def _llm_type(self) -> str:
        return "doubao"

# åˆ›å»ºLLMå®ä¾‹
llm = DoubaoLLM()

# åˆ›å»ºæç¤ºæ¨¡æ¿
prompt = PromptTemplate(
    input_variables=["question"],
    template="è¯·å›ç­”ä¸‹é¢çš„é—®é¢˜ï¼š{question}"
)

# åˆ›å»ºé“¾
chain = LLMChain(llm=llm, prompt=prompt)

# ä½¿ç”¨ç¤ºä¾‹
response = chain.run("å¸¸è§çš„åå­—èŠ±ç§‘æ¤ç‰©æœ‰å“ªäº›ï¼Ÿ")
print("é€šè¿‡Chainè°ƒç”¨çš„ç»“æœï¼š")
print(response)
print("\n")

# ç›´æ¥ä½¿ç”¨LLM
direct_response = llm("è¯·ä»‹ç»ä¸€ä¸‹äººå·¥æ™ºèƒ½")
print("ç›´æ¥è°ƒç”¨LLMçš„ç»“æœï¼š")
print(direct_response)


# del os.environ["å˜é‡å"]
```
